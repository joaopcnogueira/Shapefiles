 ### Importing the module shapefile that is contained at pyshp package

```python
import matplotlib.pyplot as plt
```

### Loading the shapefile

```python
sf = shapefile.Reader('./fortal118.shp')
```

### Reading the fields of the shapefile

```python
fields = [field[0] for field in sf.fields[1:]]
fields
```

### Accessing data from shapefile records - printing the neighborhood and its population

```python
for shape in sf.shapeRecords():
    ind_nome = fields.index('NOME')
    ind_pop = fields.index('POP')
    
    bairro = shape.record[ind_nome]
    population = shape.record[ind_pop]
    
    print(bairro, population)
```

### Ploting the shapes within the shapefile

```python
import matplotlib.pyplot as plt
shp = sf.shapeRecords()
for i in range(len(shp)):
    points = shp[i].shape.points
    lon = []
    lat = []
    for j in range(len(points)):
        lon.append(points[j][0])
        lat.append(points[j][1])
    plt.plot(lon,lat)
plt.show()
```
